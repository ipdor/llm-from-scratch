## 2.3 Converting tokens to token IDs

````python
# From processed text to vocabulary
all_words = sorted(set(preprocessed))
vocab_size = len(all_words)
print(len(preprocessed))
print(vocab_size)

vocab = {token: id for id, token in enumerate(all_words) }
print(all_words)
print(vocab)
````

````python
class SimpleTokenizerV1:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {id:token for token, id in vocab.items()}

    def encode(self, text):
        result = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        result = [item.strip() for item in result if item.strip()]
        tokenids = [self.str_to_int[token] for token in result]
        return tokenids

    def decode(self, token_ids):
        tokens = [self.int_to_str[id] for id in token_ids]
        text = " ".join(tokens)
        # 去除标点符号token前面的空格
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
````

````python
text = '''"It's the last he painted, you know," Mrs. Gisburn said with pardonable pride.'''

tkzer = SimpleTokenizerV1(vocab)
print(tkzer.str_to_int)
print(tkzer.int_to_str)

token_ids = tkzer.encode(text)
print(token_ids)

text = tkzer.decode(token_ids)
print(text)
````

````python
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {id:token for token, id in vocab.items()}

    def encode(self, text):
        result = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        result = [item.strip() for item in result if item.strip()]
        result = [
            item if item in self.str_to_int
            else '<|unk|>' for item in result
        ]
        
        tokenids = [self.str_to_int[token] for token in result]
        return tokenids

    def decode(self, token_ids):
        tokens = [self.int_to_str[id] for id in token_ids]
        text = " ".join(tokens)
        # 去除标点符号token前面的空格
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
````

````python
tkzer = SimpleTokenizerV2(vocab)
text = "Hello, do you like tea? is this-- a test?"
ids = tkzer.encode(text)
print(ids)

print(tkzer.decode(ids))
````

````python
from torch.utils.data import Dataset, DataLoader

# 作用是整合tokenizer，记录当前总训练输入和总输入+输出
# max_length是最大窗口长度，stride是每次移动多少
class GPTDatasetV1(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):
        self.input_ids = []
        self.target_ids = []

        token_ids = tokenizer.encode(txt, allowed_special={"<|endoftext|>"})

        for i in range(0, len(token_ids) - max_length, stride):
            input_chunk = token_ids[i:i+max_length]
            target_chunk = token_ids[i+1: i+max_length+1] # 1是每次加载多少作为target
            self.input_ids.append(torch.tensor(input_chunk)) 
            self.target_ids.append(torch.tensor(target_chunk))
            #print(self.input_ids)
            #print(self.target_ids)

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self,idx):
        return self.input_ids[idx], self.target_ids[idx]
````

````python
# drop_last丢弃最后一组，可能不满batch_size；
def create_dataloader_v1(txt, batch_size=2, max_length=256,
                        stride=128, shuffle=True, drop_last=True,
                        num_workers=2):
                        
    tokenizer = tiktoken.get_encoding("gpt2")

    dataset = GPTDatasetV1(txt, tokenizer, max_length, stride)

    dataLoader = DataLoader(
                dataset,
                batch_size=batch_size,
                shuffle = shuffle,
                drop_last = drop_last,
                num_workers = num_workers
    )

    return dataLoader
````

````python
with open("the-verdict.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
````

````python
dataloader = create_dataloader_v1(raw_text, batch_size=1, max_length=4, stride=1, shuffle=False,num_workers=0)

data_iter = iter(dataloader)
first_batch = next(data_iter)
print(first_batch)
````

一个简单的查找表，存储固定字典和大小的嵌入。

* embedding_dim (int) 每个嵌入向量的大小

嵌入层本质上就是一个查找操作 (lookup operation)，它通过 token ID 从嵌入层的权重矩阵中检索出对应的行。

## 2.8 Position embedding

| 问题 (Question) | 回答 (Answer) |
| :--- | :--- |
| **为什么要把文本转换成数字 ID (Token)？** | 计算机无法处理字符，只能运算数字。Tokenization 为每个词分配“身份证号”，将其转化为离散的整数索引。🔢 |
| **为什么有了 ID 还要转换成向量 (Embedding)？** | 整数 ID 本身不包含语义联系。Embedding 将 ID 映射到空间坐标，使得**语义相近**的词在空间中距离更近，捕捉词汇间的逻辑关联。📍 |
| **为什么维度一定要非常高（如 768 维），3 或 4 维不行吗？** | 人类语言极其复杂。一个词通常包含多种属性（如：苹果既是水果也是科技公司）。低维度空间无法同时满足多个语义维度的排列要求，会导致**语义冲突**或**歧义**。多维度提供了足够的“余地”来容纳复杂的词汇关系。🌌 |
|**为什么要用Position embedding**|因为模型看到的token是数学上的集合，没有顺序关系，它只计算token之间的相似度，不考虑第几个词/前后关系/顺序编号。把“位置”变成向量，加进 token embedding 里可以让模型强制考虑相对关系。考虑绝对举例->token之间的距离|
