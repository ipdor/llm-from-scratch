Exercise 2.1 Byte pair encoding of unknown words

```python
import tiktoken
```

```python
words = "Akwirw ier"
tknizer = tiktoken.get_encoding("gpt2")
tokens = tknizer.encode(words)
print(tokens)
```

```python
for tk in tokens:
    print(f'"{tknizer.decode([tk])}"', end=", ")
```

```python
decoded = tknizer.decode(tokens)
print(decoded)
```

2.7 创建 token 嵌入

```python
import torch
```

```python
input_ids = torch.tensor([2,3,5,1])
```

```python
vocab_size = 6
output_dim = 3
```

```python
torch.manual_seed(123)
embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
print(embedding_layer.weight)
```

In other words, the embedding layer is essentially a lookup operation that retrieves rows from the embedding layer’s weight matrix via a token ID.

嵌入层本质上就是一个查找操作 (lookup operation)，它通过 token ID 从嵌入层的权重矩阵中检索出对应的行。

```python
print(embedding_layer(torch.tensor([2])))
```

```python
print(embedding_layer(torch.tensor([5])))
```

```python
print(embedding_layer(torch.tensor([2, 3, 5, 1])))
```
